#Question 1
## **Basic Line Plot**
### Plot a simple line graph of ùë¶ = ùë•2 for ùë• values ranging from -10 to 10.
###Label the x-axis as "X Values" and the y-axis as "Squared Values."
###Add a title to the graph.

```python
x=range(-10,11)
y=[i**2 for i in x]
plt.plot(x,y)
plt.xlabel("X Values")
plt.ylabel("Squared Values")
plt.title("Basic Line Plot")
plt.show()
```

#Question 5
##**Pie Chart**
###Visualize the percentage distribution of five categories ['A', 'B', 'C', 'D', 'E'] with values [20, 30, 25, 15, 10].
###Display the percentage values on the chart and use a color palette for differentiation.

```python
c=['A', 'B', 'C', 'D', 'E']
v=[20, 30, 25, 15, 10]
plt.pie(
    v,
    labels=c,
    autopct='%1.0f%%',
    colors=['red', 'green', 'blue', 'yellow', 'orange']
)
plt.title("Pie Chart")
plt.show()
```
